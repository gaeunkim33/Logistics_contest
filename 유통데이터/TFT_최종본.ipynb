```python
import pandas as pd

tdf = target_df.copy()
tdf['날짜'] = pd.to_datetime(dict(year=tdf['연도'], month=tdf['월'], day=1))

start, end = pd.Timestamp('2021-01-01'), pd.Timestamp('2024-12-01')
all_months = pd.date_range(start, end, freq='MS')
expected_n = len(all_months)

cats = tdf['소분류'].astype(str).unique()
full = pd.MultiIndex.from_product([cats, all_months], names=['소분류','날짜']).to_frame(index=False)

present = tdf[['소분류','날짜']].drop_duplicates().assign(present=1)

chk = (full.merge(present, on=['소분류','날짜'], how='left')
            .assign(missing=lambda x: x['present'].isna())
            .drop(columns='present'))

summary = (chk.groupby('소분류')['missing']
             .agg(missing_months=lambda s: int(s.sum()),
                  total_months='size')
             .reset_index())
summary['expected_months'] = expected_n
summary['ok_no_gap'] = summary['missing_months'].eq(0)

missing_detail = chk.loc[chk['missing'], ['소분류','날짜']].sort_values(['소분류','날짜'])

print(f"[기간 내 총 월수 기대치] {expected_n}")
print("\n[소분류별 누락 요약] (상위 20건)")
print(summary.sort_values(['missing_months','소분류'], ascending=[False, True]).head(20))

print("\n[누락 상세 예시] (상위 20행)")
print(missing_detail.head(20))
```

```python
import pandas as pd

tdf = target_df.copy()
ds  = dataset.copy()

def _expand(row):
    s = str(row['DATE'])
    if 'Q' in s:
        y, q = s.split('-')[0], s.split('-')[1].replace('Q','')
        y, q = int(y), int(q)
        mlist = {1:[1,2,3], 2:[4,5,6], 3:[7,8,9], 4:[10,11,12]}[q]
        return pd.DataFrame({'연도':[y]*3, '월':mlist, 'NAME':[row['NAME']]*3, 'VALUE':[row['VALUE']]*3})
    else:
        y, m = s.split('-')[0], s.split('-')[1]
        y, m = int(y), int(m)
        return pd.DataFrame({'연도':[y], '월':[m], 'NAME':[row['NAME']], 'VALUE':[row['VALUE']]})

expanded = pd.concat(ds.apply(_expand, axis=1).tolist(), ignore_index=True)

# 같은 (연도,월,NAME)에 중복이 있으면 평균으로 집계(필요시 'last' 등으로 변경)
expanded = (expanded.groupby(['연도','월','NAME'], as_index=False)['VALUE']
                    .mean())

# 월단위 wide 형태로 피벗 → NAME들이 컬럼이 됨
macro_wide = (expanded.pivot(index=['연도','월'], columns='NAME', values='VALUE')
                        .reset_index())

# target_df에 병합
target_df = tdf.merge(macro_wide, on=['연도','월'], how='left')
```

```python
# 월별 주말/공휴일/요일별 개수를 계산해 target_df에 추가
import pandas as pd
import numpy as np
from calendar import monthrange
from holidayskr.core import is_holiday  # pip install holidayskr

# (연도, 월) 유니크 키
keys = (
    target_df[['연도','월']]
    .drop_duplicates()
    .astype({'연도': int, '월': int})
    .reset_index(drop=True)
)

def month_stats(year: int, month: int) -> dict:
    # 해당 월의 모든 날짜 생성
    n_days = monthrange(year, month)[1]
    dates = pd.date_range(f'{year}-{month:02d}-01', periods=n_days, freq='D')

    # 요일별 개수 (월=0, ..., 일=6)
    wd_cnt = np.bincount(dates.weekday, minlength=7)
    mon, tue, wed, thu, fri, sat, sun = (int(wd_cnt[i]) for i in range(7))

    # 주말일수(토+일)
    weekends = sat + sun

    # 공휴일 개수(문자열로 넘겨야 라이브러리에서 처리됨)
    holidays = int(sum(is_holiday(d.strftime('%Y-%m-%d')) for d in dates))

    return {
        '월요일수': mon, '화요일수': tue, '수요일수': wed, '목요일수': thu,
        '금요일수': fri, '토요일수': sat, '일요일수': sun,
        '주말일수': weekends, '공휴일수': holidays
    }

# 월별 집계 테이블 생성
rows = []
for y, m in keys.itertuples(index=False):
    rec = {'연도': int(y), '월': int(m)}
    rec.update(month_stats(int(y), int(m)))
    rows.append(rec)

cal_df = pd.DataFrame(rows)

# target_df에 병합
target_df = target_df.merge(cal_df, on=['연도','월'], how='left')
```

```python
# ============================================================
# TFT 월별 예측 파이프라인 (학습=2021-01~2023-12, 검증=2024-01~12, 예측=2025-01~03)
#  - 학습: 전체 소분류 (예측 대상: 봉지라면/탄산음료)
#  - 인코더 24개월 / 디코더 3개월
#  - 날씨: 월평균(평균/최저/최고 모두 '월 평균')
#  - 거시지표: YYYY-Qn → 분기 내 3개월 동일 복제
#  - 전제: target_df(2021-01~2024-12), dataset[NAME,DATE,VALUE](2025-01~03 포함),
#          weather_df(일단위)가 이미 로드되어 있음
# ============================================================

import pandas as pd, numpy as np, re
import torch as T
import lightning.pytorch as pl
from pytorch_forecasting import TimeSeriesDataSet, TemporalFusionTransformer
from pytorch_forecasting.data import GroupNormalizer
from pytorch_forecasting.metrics import QuantileLoss
from lightning.pytorch.callbacks import ModelCheckpoint, EarlyStopping

# -----------------------
# 0) 설정
# -----------------------
ENC_LEN = 24
PRED_LEN = 3                      # 2025-01~03
LR = 3e-4
BATCH_SIZE = 64
MAX_EPOCHS = 50
GROUP_COL   = "소분류"
TARGET_COL  = "target"
YEAR_COL    = "연도"
MONTH_COL   = "월"
VAL_START   = pd.Timestamp("2024-01-01")   # 검증 시작
FORECAST_MONTHS = pd.period_range("2025-01", "2025-03", freq="M").to_timestamp()
TARGET_CATS = ["봉지라면", "탄산음료"]  # 예측 대상 순서 고정

pl.seed_everything(42)

# ----------------------------------------------------
# 1) 날짜 표준화 (월초 timestamp: date)
# ----------------------------------------------------
def ensure_ym(df: pd.DataFrame) -> pd.Series:
    if "date" in df.columns:
        dt = pd.to_datetime(df["date"], errors="coerce")
    elif "ds" in df.columns:
        dt = pd.to_datetime(df["ds"], errors="coerce")
    elif {"연도","월"}.issubset(df.columns):
        dt = pd.to_datetime(df["연도"].astype(int).astype(str)+"-"+df["월"].astype(int).astype(str)+"-01", errors="coerce")
    elif "날짜" in df.columns:
        dt = pd.to_datetime(df["날짜"], errors="coerce")
    else:
        raise KeyError("월 기준 날짜 컬럼을 만들 수 없습니다. (연도/월 or ds or 날짜 필요)")
    if dt.isna().any():
        raise ValueError("날짜 파싱 중 NaT가 발생했습니다.")
    return dt.dt.to_period("M").dt.to_timestamp()

target_df = target_df.copy()
target_df["date"] = ensure_ym(target_df)

# ----------------------------------------------------
# 2) 날씨: 일→월 평균 집계 (평균/최저/최고 모두 '월 평균')
# ----------------------------------------------------
def prepare_weather_monthly(weather_df: pd.DataFrame) -> pd.DataFrame:
    w = weather_df.copy()
    base_col = "날짜" if "날짜" in w.columns else ("date" if "date" in w.columns else None)
    if base_col is None:
        raise KeyError("weather_df에 '날짜' 또는 'date' 컬럼이 필요합니다.")
    w[base_col] = w[base_col].astype(str).str.strip()
    w["date"] = pd.to_datetime(w[base_col], errors="coerce")
    w = w.dropna(subset=["date"])
    w["ym"] = w["date"].dt.to_period("M").dt.to_timestamp()

    need = {"평균기온(℃)":"월평균기온(℃)","최저기온(℃)":"월최저기온(℃)","최고기온(℃)":"월최고기온(℃)"}
    missing = [c for c in need if c not in w.columns]
    if missing: raise KeyError(f"weather_df에 필요 컬럼 없음: {missing}")

    m = (w.groupby("ym", as_index=False)
           .agg({k:"mean" for k in need})
           .rename(columns={"ym":"date", **need}))
    return m

weather_m = prepare_weather_monthly(weather_df)

# ----------------------------------------------------
# 3) 거시지표 전개 (YYYY-Qn/ YYYY-MM → 월별 확장, NAME pivot)
# ----------------------------------------------------
_qpat = re.compile(r"^(\d{4})-Q([1-4])$")
_mpat = re.compile(r"^(\d{4})-(\d{2})$")

def expand_macro(dataset: pd.DataFrame) -> pd.DataFrame:
    if not {"NAME","DATE","VALUE"}.issubset(set(map(str,dataset.columns))):
        raise KeyError("dataset은 NAME, DATE, VALUE 컬럼 필요")
    d = dataset.copy()
    rows = []
    for name, date_s, val in d[["NAME","DATE","VALUE"]].itertuples(index=False):
        s = str(date_s).strip()
        qm = _qpat.match(s); mm = _mpat.match(s)
        if qm:
            y = int(qm.group(1)); q = int(qm.group(2)); start_m = (q-1)*3+1
            for m in [start_m, start_m+1, start_m+2]:
                rows.append([name, pd.Timestamp(y, m, 1), float(val)])
        elif mm:
            y = int(mm.group(1)); m = int(mm.group(2))
            rows.append([name, pd.Timestamp(y, m, 1), float(val)])
        else:
            continue
    ex = pd.DataFrame(rows, columns=["NAME","date","VALUE"])
    wide = ex.pivot_table(index="date", columns="NAME", values="VALUE", aggfunc="last").reset_index()
    return wide

macro_wide = expand_macro(dataset)

# ----------------------------------------------------
# 4) 2025-01~03 미래 피처 생성 (두 소분류 × 3개월)
# ----------------------------------------------------
def make_month_cal(df_like: pd.DataFrame, months: pd.DatetimeIndex) -> pd.DataFrame:
    needed = ["월요일수","화요일수","수요일수","목요일수","금요일수","토요일수","일요일수","주말일수","공휴일수"]
    present = [c for c in needed if c in df_like.columns]
    out = []
    for ts in months:
        dates = pd.date_range(ts, ts+pd.offsets.MonthEnd(0), freq="D")
        dow = dates.weekday
        cnt = {
            "월요일수":int((dow==0).sum()), "화요일수":int((dow==1).sum()), "수요일수":int((dow==2).sum()),
            "목요일수":int((dow==3).sum()), "금요일수":int((dow==4).sum()), "토요일수":int((dow==5).sum()),
            "일요일수":int((dow==6).sum())
        }
        cnt["주말일수"] = cnt["토요일수"]+cnt["일요일수"]
        cnt["공휴일수"] = 0
        out.append({"date":ts, "연도":ts.year, "월":ts.month, **{k:cnt[k] for k in present}})
    return pd.DataFrame(out)

def make_future_features(base_like: pd.DataFrame,
                         months: pd.DatetimeIndex,
                         macro_wide: pd.DataFrame,
                         weather_m: pd.DataFrame) -> pd.DataFrame:
    base = pd.DataFrame({GROUP_COL: TARGET_CATS}).assign(key=1)
    mon  = pd.DataFrame({"date": months, "key":1})
    grid = base.merge(mon, on="key").drop(columns="key")
    cal  = make_month_cal(df_like=base_like, months=months)
    fut  = (grid.merge(cal, on="date", how="left")
                 .merge(macro_wide, on="date", how="left")
                 .merge(weather_m,  on="date", how="left"))
    fut[TARGET_COL] = np.nan
    order = [GROUP_COL, TARGET_COL, YEAR_COL, MONTH_COL] + \
            [c for c in base_like.columns if c not in [GROUP_COL,TARGET_COL,YEAR_COL,MONTH_COL,"date"]]
    cols = [c for c in order if c in fut.columns] + [c for c in fut.columns if c not in order]
    return fut[cols].copy()

future_2025 = make_future_features(target_df, FORECAST_MONTHS, macro_wide, weather_m)

# ----------------------------------------------------
# 5) full_df + time_idx
# ----------------------------------------------------
full_df = target_df.copy()
all_months = pd.period_range(full_df["date"].min(), full_df["date"].max(), freq="M").to_timestamp()
month_to_idx = {d:i for i,d in enumerate(all_months)}
full_df["time_idx"] = full_df["date"].map(month_to_idx)

# ----------------------------------------------------
# 6) 변수 정의
# ----------------------------------------------------
static_categoricals = [GROUP_COL]
time_varying_known_reals = []
time_varying_unknown_reals = [TARGET_COL]

if YEAR_COL in full_df.columns:  time_varying_known_reals.append(YEAR_COL)
if MONTH_COL in full_df.columns: time_varying_known_reals.append(MONTH_COL)

exclude = {GROUP_COL, TARGET_COL, "date", "time_idx", YEAR_COL, MONTH_COL}
num_cols = [c for c in full_df.columns if c not in exclude and pd.api.types.is_numeric_dtype(full_df[c])]
time_varying_known_reals += num_cols

# ----------------------------------------------------
# 7) 학습/검증 데이터셋
# ----------------------------------------------------
val_start_idx   = month_to_idx[VAL_START]
training_cutoff = val_start_idx - 1  # 2023-12

training = TimeSeriesDataSet(
    full_df[full_df["time_idx"] <= training_cutoff],
    time_idx="time_idx",
    target=TARGET_COL,
    group_ids=[GROUP_COL],
    min_encoder_length=ENC_LEN, max_encoder_length=ENC_LEN,
    min_prediction_length=PRED_LEN, max_prediction_length=PRED_LEN,
    static_categoricals=static_categoricals,
    time_varying_known_reals=time_varying_known_reals,
    time_varying_unknown_reals=time_varying_unknown_reals,
    target_normalizer=GroupNormalizer(groups=[GROUP_COL]),
    add_relative_time_idx=True, add_target_scales=True, add_encoder_length=True,
)

validation = TimeSeriesDataSet.from_dataset(
    training, full_df, predict=True, stop_randomization=True,
    min_prediction_idx=val_start_idx
)

train_loader = training.to_dataloader(train=True,  batch_size=BATCH_SIZE, num_workers=0)
val_loader   = validation.to_dataloader(train=False, batch_size=BATCH_SIZE, num_workers=0)

# ----------------------------------------------------
# 8) 학습
# ----------------------------------------------------
ckpt_cb = ModelCheckpoint(monitor="val_loss", save_top_k=1, mode="min")
es_cb   = EarlyStopping(monitor="val_loss", patience=5, mode="min")

tft = TemporalFusionTransformer.from_dataset(
    training,
    learning_rate=LR, hidden_size=32, attention_head_size=4, dropout=0.1,
    loss=QuantileLoss(), log_interval=50, reduce_on_plateau_patience=3,
)

trainer = pl.Trainer(
    max_epochs=MAX_EPOCHS, accelerator="auto", gradient_clip_val=1.0,
    callbacks=[ckpt_cb, es_cb], enable_progress_bar=True,
)

trainer.fit(tft, train_loader, val_loader)

best_path = ckpt_cb.best_model_path
best_tft  = TemporalFusionTransformer.load_from_checkpoint(best_path)

# ----------------------------------------------------
# 9) 검증 지표: val_loss 출력
# ----------------------------------------------------
val_metrics = trainer.validate(best_tft, dataloaders=val_loader, verbose=False)
if len(val_metrics) > 0 and "val_loss" in val_metrics[0]:
    print(f"[VAL] val_loss={val_metrics[0]['val_loss']:.4f}")
else:
    print("[VAL] validation metrics unavailable.")

# ----------------------------------------------------
# 10) 2025-01~03 예측 (히스토리 포함 + 미래 target 더미값 채움)
# ----------------------------------------------------
# time_idx 확장
last_idx = max(month_to_idx.values())
for ts in FORECAST_MONTHS:
    if ts not in month_to_idx:
        last_idx += 1
        month_to_idx[ts] = last_idx

# 미래 프레임에 time_idx 부여
future_2025 = future_2025.copy()
future_2025["date"] = future_2025["date"].dt.to_period("M").dt.to_timestamp()
future_2025["time_idx"] = future_2025["date"].map(month_to_idx)

# 대상 소분류의 최근 24개월 히스토리(인코더용)
last_train_idx = full_df["time_idx"].max()       # 2024-12 index
hist_start_idx = last_train_idx - ENC_LEN + 1    # 최근 24개월 시작
hist = full_df[
    (full_df[GROUP_COL].isin(TARGET_CATS)) &
    (full_df["time_idx"] >= hist_start_idx)
].copy()

# 스키마 맞추기
train_like_cols = [c for c in full_df.columns if c != "date"]
future_aligned = future_2025.copy()
for col in train_like_cols:
    if col not in future_aligned.columns:
        future_aligned[col] = np.nan
future_aligned = future_aligned[train_like_cols]
hist = hist[train_like_cols]

# 히스토리 + 미래 합치기
future_with_hist = pd.concat([hist, future_aligned], ignore_index=True)

# ★ 미래 구간 target NaN → 더미값(0.0) 채우기 (TFT 버전 호환)
forecast_start_idx = month_to_idx[min(FORECAST_MONTHS)]
future_mask = future_with_hist["time_idx"] >= forecast_start_idx
future_with_hist.loc[future_mask, TARGET_COL] = 0.0

# 예측에 쓰이는 샘플의 그룹 순서(안정성 위해 우리가 명시)
group_order = TARGET_CATS[:]  # ["봉지라면","탄산음료"]

# 안전: 히스토리 구간도 이 순서로 정렬
future_with_hist = future_with_hist.sort_values([GROUP_COL, "time_idx"], key=lambda s: s if s.name!="소분류" else s.map({g:i for i,g in enumerate(group_order)})).reset_index(drop=True)

# 예측용 데이터셋/로더
pred_ds = TimeSeriesDataSet.from_dataset(
    training, future_with_hist, stop_randomization=True, predict=True
)
pred_loader = pred_ds.to_dataloader(train=False, batch_size=BATCH_SIZE, num_workers=0)

# ----------------------------------------------------
# 11) 예측 실행 (가장 호환성 높은 point 예측만: p50)
# ----------------------------------------------------
# 반환은 [N, PRED_LEN] (텐서 또는 넘파이)
pred_point = best_tft.predict(pred_loader, mode="prediction")
pred_np = pred_point.detach().cpu().numpy() if hasattr(pred_point, "detach") else np.asarray(pred_point)

# N = 그룹 수(=2)라고 가정. (우리가 입력을 그룹 순서 고정으로 만들었기 때문)
N = pred_np.shape[0]
assert N == len(group_order), f"예상 샘플 수({len(group_order)})와 예측 출력 N({N})가 다릅니다."

rows = []
for i, g in enumerate(group_order):
    for j, dt in enumerate(FORECAST_MONTHS):
        rows.append({
            GROUP_COL: g,
            "date": dt,
            "p50": float(pred_np[i, j]),
        })

pred_out = pd.DataFrame(rows).sort_values([GROUP_COL, "date"]).reset_index(drop=True)
print(pred_out)

# ----------------------------------------------------
# 12) 저장
# ----------------------------------------------------
pred_out.to_csv("tft_predictions_2025Q1.csv", index=False)
print("Saved -> tft_predictions_2025Q1.csv")
```
